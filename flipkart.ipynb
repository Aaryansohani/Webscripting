```python
import os
import numpy as np
import pandas as pd
import datetime
import time
import dateutil
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
from bs4 import BeautifulSoup as bs
import warnings 
warnings.filterwarnings('ignore')
```

# Crawler


```python
options = Options()
options.add_argument('--start-maximized')
#options.add_argument('--headless')
driver = webdriver.Chrome(options = options)

#link
driver.get('https://www.flipkart.com/')
time.sleep(5)

#login prompt
driver.find_element_by_xpath('/html/body/div[2]/div/div/button').click()
time.sleep(2)

#click on searching
driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').send_keys('samsung mobiles')

# submiting
driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()

time.sleep(5)

#inserting page1
page1 = driver.page_source
time.sleep(5)

#page 2
driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[2]/div[1]/div[2]/div[26]/div/div/nav/a[2]').click()
page2 = driver.page_source
time.sleep(5)

driver.quit()
```

```python
page1
```

```python
#putting the fetched data into beautifulsoup
soup = bs(page1,'html.parser')
```

```python
#removing noise
col1 = soup.findAll('div',{'class' : '_3wU53n'})
col2 = soup.findAll('div',{'class' : '_1vC4OE _2rQ-NK'})
col2
```

```python
# Fetching data of phones
productname = [x.text for x in col1] 
productname
```

```python
# Fetching data of prices
product_price = [x.text for x in col2]
product_price
```

```python
# joing both the lists
data_list = [productname,product_price]
data_list
```

```python
# Converting in a dataframe
df = pd.DataFrame(data = data_list)
df
```

```python
# Transposing DataFrame
df = df.transpose()
```

```python
#renaming the columns
df.rename(columns ={0 :'Phones',1 : 'Price'})
```

```python
#exporting as csv
df.to_csv('flipkart.csv', index=False)
```
